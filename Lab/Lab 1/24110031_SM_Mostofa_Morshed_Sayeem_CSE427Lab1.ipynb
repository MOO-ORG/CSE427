## NumPy

```python
import numpy as np
```

### Task 1: Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.

```python
a = np.random.randint(1, 101, size=20)
b = np.random.randint(1, 101, size=20)

print("Array a:", a)
print("Array b:", b)
```

### Task 2: Reshape both linear arrays into two 5x4 matrices.

```python
a = a.reshape(5, 4)
b = b.reshape(5, 4)

print("Matrix a:")
print(a)
print("\nMatrix b:")
print(b)
```

### Task 3: Perform scalar multiplication with matrix a and b.

```python
scalar = 3

print("Matrix a multiplied by", scalar, ":")
print(a * scalar)
print("\nMatrix b multiplied by", scalar, ":")
print(b * scalar)
```

### Task 4: Modify matrix b and perform matrix multiplication of a and b, storing the result in c.

Matrix `a` is 5x4. For matrix multiplication to be valid, the number of columns of `a` (4) must equal the number of rows of `b`. Matrix `b` is reshaped to 4x5 so the product is defined. The result `c` is 5x5.

```python
b = b.reshape(4, 5)
c = a @ b

print("Modified matrix b (4x5):")
print(b)
print("\nMatrix c = a @ b (5x5):")
print(c)
```

### Task 5: Print only columns 2 and 3 of matrix c. Then print the whole matrix using slicing.

```python
print("Columns 2 and 3 of matrix c:")
print(c[:, 1:3])

print("\nWhole matrix c using slicing:")
print(c[:, :])
```

### Task 6: Find the largest element in the array and the index of that element.

```python
largest = c.max()
flat_index = c.argmax()
position = np.unravel_index(flat_index, c.shape)

print("Largest element:", largest)
print("Flattened index:", flat_index)
print("Position (row, column):", position)
```

### Task 7: Convert matrix c into a one dimensional array while keeping the original matrix unchanged.

```python
c_flat = c.flatten()

print("One dimensional array:")
print(c_flat)
print("\nOriginal matrix c (unchanged):")
print(c)
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

### Task 1: Scatter plot of temperature_dhaka vs humidity_dhaka using (*) as the marker in red.

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

plt.scatter(temperature_dhaka, humidity_dhaka, marker="*", color="red")
plt.xlabel("Temperature (Dhaka)")
plt.ylabel("Humidity (Dhaka)")
plt.title("Temperature vs Humidity")
plt.show()
```

### Task 2: Line plot of study_hours vs marks with figure size (12, 8).

```python
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

### Task 3: Horizontal bar graph showing the scores in each subject.

```python
subjects = ["Maths", "English", "Science", "Physics", "Computer"]
subject_marks = [89, 90, 45, 78, 99]

plt.barh(subjects, subject_marks)
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.title("Subject Scores")
plt.show()
```

### Task 4: Vertical bar graph where each subject score is in a different color.

```python
colors = ["red", "green", "blue", "orange", "purple"]

plt.bar(subjects, subject_marks, color=colors)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Scores")
plt.show()
```

### Task 5: Show all of these graphs in one figure using the subplot function.

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker="*", color="red")
plt.xlabel("Temperature (Dhaka)")
plt.ylabel("Humidity (Dhaka)")
plt.title("Temperature vs Humidity")

plt.subplot(2, 2, 2)
plt.plot(study_hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

plt.subplot(2, 2, 3)
plt.barh(subjects, subject_marks)
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.title("Subject Scores (Horizontal)")

plt.subplot(2, 2, 4)
plt.bar(subjects, subject_marks, color=colors)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Scores (Vertical)")

plt.tight_layout()
plt.show()
```
